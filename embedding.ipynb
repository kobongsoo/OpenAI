```python
# 임베딩 예시
# 참조 : https://platform.openai.com/docs/guides/embeddings/what-are-embeddings

import os
import openai

# key 지정
openai.api_key = "sk-xxx"

# 모델 - embeding 모델 지정
# => 모델 목록은 : https://platform.openai.com/docs/models/gpt-4 참조
model = "text-embedding-ada-002"

text = "오늘은 날씨가 너무 좋다."
response = openai.Embedding.create(
    input=text,
    model=model
)
embeddings = response['data'][0]['embedding']

print(f'len:{len(embeddings)}')
print(embeddings)
```
